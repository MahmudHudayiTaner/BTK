```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
yasListesi = [10,20,30,40,50,60,70,80]
kiloListesi = [20,60,80,85,86,87,70,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyYasListesi
```

```python
numpyKiloListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi, "g")
plt.xlabel("yas")
plt.ylabel("kilo")
plt.title("Kilo'nun Yaşa Göre Değişim Grafiği")
plt.show()
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g")
plt.show()
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
plt.show()
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g*-")
plt.show()
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"b-*")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r")
plt.subplot(1,2,2)
plt.plot(numpyDizisi2,numpyDizisi1,"b--")
plt.show()
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.6,0.6,0.5,0.5]) #( , , width = genişlik,height=yükseklik)
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
plt.show()
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0,0,0.7,0.7])
eksen2 = figure2.add_axes([0.2,0.1,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")

plt.show()
```

```python
plt.figure()
plt.show()
```

```python
plt.subplots()
plt.show()
```

```python
(benimFigure,benimEksen) = plt.subplots(nrows = 1, ncols = 2)

for eksen in benimEksen:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    
plt.show()
```

```python
yeniFigure = plt.figure()

yeniEksen = yeniFigure.add_axes([0,0,0.5,0.5])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2,"g",label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 3,"r", label = "numpyDizisi ** 3")

yeniEksen.legend( loc = 4)

plt.show()
```

```python
yeniFigure.savefig("örnekFigure.png",dpi = 200)
```

```python
(stilFigure,stilEksen) = plt.subplots()

stilEksen.plot(numpyDizisi1,numpyDizisi1 + 2, color="blue", linewidth = 1)

stilEksen.plot(numpyDizisi1,numpyDizisi1 + 3, color="r", linewidth = 1)

stilEksen.plot(numpyDizisi1,numpyDizisi1 + 4, color="black", linestyle = "-.")

stilEksen.plot(numpyDizisi1,numpyDizisi1 + 5, color="black", linestyle = "-")

stilEksen.plot(numpyDizisi1,numpyDizisi1 + 6, color="black", linestyle="--")

stilEksen.plot(numpyDizisi1,numpyDizisi1 + 7, color="black", ls=":")

plt.show()
```
